# 🔷 Description générale du projet

**EasySave** est une application de sauvegarde de fichiers et de répertoires, développée pour répondre aux besoins d’utilisateurs souhaitant **sécuriser et automatiser** leurs copies de données, que ce soit sur des supports :

- Locaux (disques internes),
- Externes (clés USB, disques durs),
- Ou en réseau.

---

## Fonctionnalités principales

L’application permet de **créer et gérer jusqu’à 5 travaux de sauvegarde**, chacun défini par :

- Un **nom personnalisé**,
- Un **répertoire source** (à sauvegarder),
- Un **répertoire cible** (de destination),
- Un **type de sauvegarde** :
  - **Complète** (copie intégrale),
  - **Différentielle** (copie des fichiers modifiés depuis la dernière sauvegarde complète).

L’utilisateur peut **exécuter un ou plusieurs travaux** via une **interface en ligne de commande**, avec une **syntaxe simple et efficace**.

---

## Suivi et traçabilité

**EasySave** génère automatiquement **deux fichiers** au format **JSON** pour assurer le suivi des opérations :

1. **Journal d’exécution (log)** :  
   Trace chaque fichier traité (nom, taille, durée de transfert…).

2. **Fichier d’état** :  
   Indique en temps réel la progression de chaque sauvegarde (nombre de fichiers restants, volume restant, etc.).

Ces fichiers sont :

- **Lisibles** et directement exploitables via un **éditeur de texte simple** (type Notepad),
- **Formatés** avec des retours à la ligne pour chaque élément,
- **Pensés pour une utilisation facile sur les serveurs clients**.

## 🔹 Étape 1 : Diagramme de cas d'utilisation – *EasySave v1.0*

Le diagramme de cas d'utilisation ci-dessous représente les **interactions principales entre l’utilisateur et le logiciel EasySave**, dans sa version **console**.  
Il constitue une base essentielle pour la **modélisation fonctionnelle** du système, en identifiant les **fonctionnalités accessibles à l’utilisateur**.



---

### 👤 Acteur : Utilisateur

L’acteur principal est l’**Utilisateur**, qui interagit avec le logiciel pour **créer et exécuter des travaux de sauvegarde**, en ligne de commande.  
Il peut être **francophone ou anglophone**, conformément aux exigences du cahier des charges.

---

### Cas d’utilisation principaux :

#### Créer un travail de sauvegarde : 

Ce cas d’utilisation encapsule toutes les étapes nécessaires à la définition d’un travail :

- `<<include>>` **Saisir nom** : entrée du nom du travail.
- `<<include>>` **Sélectionner source** : choix du répertoire à sauvegarder.
- `<<include>>` **Sélectionner cible** : choix du répertoire de destination.
- `<<include>>` **Choisir type** : sélection du type de sauvegarde, qui peut conduire à :
  - `<<extend>>` **Sauvegarde complète**
  - `<<extend>>` **Sauvegarde différentielle**

#### Exécuter un ou plusieurs travaux de sauvegarde : 

L’utilisateur peut déclencher l’exécution d’un ou plusieurs travaux :

- `<<include>>` **Écrire dans le fichier de log** : enregistrement en temps réel des actions dans un journal au format JSON.

---

### ✅ Remarques :

- Un **doublon visuel** est présent pour le cas *"Exécuter un ou plusieurs travaux de sauvegarde"* et pourra être **corrigé ou interprété** selon la stratégie d’exécution retenue (*séquentielle vs individuelle*).


## 🔹 Étape 2 : Diagramme de séquence – *Exécution d’une sauvegarde*

Le diagramme de séquence présenté ci-dessus décrit le **scénario d’exécution d’un travail de sauvegarde** dans l’application *EasySave*.  
Il modélise les **interactions entre les différents objets (ou classes)** impliqués dans ce processus, en suivant un **enchaînement temporel de messages**.


---

### Objectif : 

Ce diagramme illustre le **déroulement complet de l’action "lancer une sauvegarde"**, depuis l’interaction de l’utilisateur jusqu’à la **copie effective des fichiers**, en passant par la **récupération des chemins** et la **gestion des logs**.

---

### Objets participants : 

- **Utilisateur** : déclenche l’action de sauvegarde.
- **UI (Interface Utilisateur)** : capte l'action de l'utilisateur et appelle le contrôleur.
- **BackupController** : pilote le processus de sauvegarde.
- **FileManager** : gère les accès aux fichiers (liste, copie).
- **ConfigManager** : gère les chemins de configuration.

---

### Séquence d’exécution : 

1. L’utilisateur clique sur **“Sauvegarder”** dans l’interface console ou graphique.
2. L’UI appelle la méthode `lancerSauvegarde()` du **BackupController**.
3. Le **BackupController** consulte le **FileManager** pour initier la sauvegarde avec les paramètres `cheminSource` et `cheminDestination`.
4. Le **FileManager** récupère la liste des fichiers à sauvegarder via la méthode `listerFichiers(cheminSource)`.
5. Pour chaque fichier de cette liste, une **boucle** est effectuée :
   - Le fichier est copié avec `copierFichier(src, dest)`.
   - Une ligne de log est générée : `log("fichier copié")`.
6. Une fois la boucle terminée :
   - Un message `log("opération terminée")` est envoyé à l’utilisateur pour signaler la fin de la sauvegarde.

---

### ✅ Remarques : 

- Le **BackupController centralise la logique métier**, ce qui respecte le principe de séparation des responsabilités.
- Les **logs sont enregistrés à chaque étape importante**, comme requis dans le cahier des charges.
- Le diagramme met en avant une **architecture modulaire et maintenable**, idéale pour les **évolutions futures du logiciel**.


## 🔹 Étape 3 : Diagramme d’activités – Parcours utilisateur dans EasySave

Ce diagramme d’activités décrit le **cheminement global de l’utilisateur** lors de l’utilisation de l’application **EasySave** en version console.  
Il permet de visualiser les **différentes étapes logiques** proposées par le programme, du lancement jusqu’à la fin de l’exécution, en passant par la **gestion des sauvegardes**.



---

### 🔁 Déroulement global : 

- L'utilisateur commence par **exécuter EasySave**, ce qui affiche un **menu principal**.
- Avant toute action, l'utilisateur doit **choisir la langue** de l’interface (*français* ou *anglais*) afin de garantir une utilisation adaptée à son profil.

Il peut ensuite choisir parmi trois options :
- Créer une nouvelle sauvegarde
- Charger une sauvegarde existante
- Quitter le programme

---

### Charger une sauvegarde

S’il choisit de charger une sauvegarde :

- Le programme affiche la **liste des sauvegardes** déjà enregistrées.
- L’utilisateur saisit le **nom de la sauvegarde** à lancer.
- Le programme exécute cette sauvegarde, tout en :
  - Écrivant les opérations dans un **fichier journalier (log)** ;
  - Mettant à jour l’**état d’avancement** dans un second fichier.

---

### Créer une sauvegarde : 

S’il choisit de créer une sauvegarde :

- Le programme vérifie si le **maximum de 5 sauvegardes** a déjà été atteint :
  - Si c’est le cas, l’utilisateur est **renvoyé au menu principal**.
  - Sinon, il peut continuer.

- L’utilisateur doit alors choisir le **type de sauvegarde** :
  - Sauvegarde **Complète**
  - Sauvegarde **Différentielle**

- Dans les deux cas, il lui est ensuite demandé :
  - Le **nom** de la sauvegarde
  - Le **chemin du répertoire source**
  - Le **chemin du répertoire de destination**

- Une fois ces informations saisies, elles sont enregistrées dans un **fichier de configuration au format JSON** (`SaveTasks.json`), permettant de les **réutiliser plus tard**.

---

###  ✅ Points importants : 

- Le logiciel **limite volontairement** le nombre de sauvegardes créées à **cinq**, ce qui nécessite une **vérification systématique** lors de chaque création.
- Le **bilinguisme** fait partie intégrante du fonctionnement, avec une interface en **français ou en anglais dès le démarrage**.
- L’**expérience utilisateur reste simple et linéaire**, avec des boucles de retour vers le menu lorsqu’une condition n’est pas remplie.
- Le format **JSON** utilisé pour stocker les sauvegardes assure une **bonne lisibilité** et une **intégration facilitée** avec d'autres composants du système.


# 🔷 Étape 4 : Diagramme de classes – Architecture objet d’EasySave

Ce diagramme de classes représente la structure interne du logiciel **EasySave** et permet de visualiser les différents composants du système, ainsi que les relations entre eux.  
Il a été conçu selon les principes de la **programmation orientée objet**, afin de garantir une organisation claire, modulaire et évolutive du code.



---

##  Organisation générale

- Le point d’entrée du programme est la classe **Program**, qui initialise le gestionnaire principal de l’application.
- La classe centrale est **ApplicationManager**, qui coordonne l’ensemble des fonctionnalités du logiciel.
- `ApplicationManager` fait appel à un **BackupManager**, responsable de la gestion des sauvegardes.

---

##  Gestion des sauvegardes

Le cœur fonctionnel est assuré par les classes suivantes :

- **BackupManager** : gère la liste des sauvegardes (`BackupJob`), les ajouts, suppressions, et exécutions.
- **BackupJob** : représente une tâche de sauvegarde. Elle contient un nom, un dossier source, un dossier de destination, et un type (complète ou différentielle).
- **BackupType** : une énumération (`enum`) qui définit les deux types de sauvegardes possibles.
- **BackupProgress** : stocke les informations sur l’avancement d’une tâche (nombre de fichiers, taille, état, etc.).
- **BackupState** : un `enum` qui indique si une sauvegarde est active ou inactive.

---

##  Suivi des actions en temps réel

Deux composants spécialisés sont utilisés pour le suivi :

- **LogManager** : enregistre les opérations effectuées dans un fichier journalier au format JSON.
  - Il utilise la classe **LogEntry** pour stocker les données à écrire (nom de fichier, taille, durée du transfert, etc.).

- **StateManager** : met à jour un fichier JSON indiquant l’état de la sauvegarde en cours.

---

##  Gestion multilingue

- La classe **LanguageManager** permet de gérer les traductions (français/anglais) grâce à un dictionnaire de chaînes.
- Elle permet d’afficher tous les menus dans la langue choisie par l’utilisateur au démarrage.

---

## ✅ Points importants

- L’architecture suit le principe **séparation des responsabilités** : chaque classe a un rôle précis.
- Le code est **modulaire et maintenable**, avec des dépendances limitées entre les composants.
- Les **logs** et **l’état** sont gérés dans deux fichiers distincts, comme exigé dans le cahier des charges.
- Le design est prêt à accueillir une **interface graphique (v2.0)**, sans modifier le cœur du logiciel.
- Le diagramme nous a permis de **planifier notre développement de manière claire et cohérente**, en anticipant les besoins futurs.
